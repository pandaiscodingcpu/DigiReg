```python
# fetching MNIST data
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784',as_frame=False)
```

```python
X,y = mnist.data,mnist.target
```

```python
X_train,X_test,y_train,y_test = X[:60000],X[60000:],y[:60000],y[60000:]
```

```python
# displaying a sample image from mnist
import matplotlib.pyplot as plt
def plot_digit(img_data):
    img = img_data.reshape(28,28)
    plt.imshow(img,cmap='binary')
    plt.axis("off")
sample_digit = X[1]
plot_digit(sample_digit)
plt.show()
```

```python
# training a multiclass classifier using SVC
from sklearn.ensemble import RandomForestClassifier
rf_cf = RandomForestClassifier(random_state=50)
rf_cf.fit(X_train,y_train)
```

```python
rf_cf.predict([sample_digit])
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(rf_cf,X_train,y_train,cv=3,n_jobs = 2)
score
```

```python
score.mean()
```

```python
import joblib
joblib.dump(rf_cf, 'digireg.joblib')
print("Model Saved Successfully")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
y_pred = rf_cf.predict(X_train)
cm = confusion_matrix(y_train, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
```
